```python
X = data2.drop('FTR',axis=1)
y = data2['FTR']
X.shape, y.shape
```

```python
C=1 #SVM regularization parameter
```

```python
#SVC Support vector classifier
svc = svm.SVC(kernel='linear',C=C).fit(X,y)
```

```python
rbf_svc = svm.SVC(kernel='rbf',gamma=0.5,C=C).fit(X,y)
```

```python
poly_svc=svm.SVC(kernel='poly',degree=5,C=C).fit(X,y)
```

```python
linear_svc=svm.LinearSVC(C=C).fit(X,y)
```

Output Prediction

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
y_pred=rbf_svc.predict(X)
```

```python
y_pred
```

```python
accuracy_score(y,y_pred)
```

```python
confusion_matrix(y,y_pred)
```

```python
classification_report(y,y_pred)
```

```python
y_pred=linear_svc.predict(X)
print("Linear Classifier-->",accuracy_score(y,y_pred))

y_pred=poly_svc.predict(X)
print("Polynomial Classifier-->",accuracy_score(y,y_pred))
```
